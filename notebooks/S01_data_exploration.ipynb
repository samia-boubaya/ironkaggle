```python
import pandas as pd
from IPython.display import display

# Load train and test sets in dataframes
train = pd.read_csv("../data/raw/train.csv")
test = pd.read_csv("../data/raw/test.csv")

# Display df_train DataFrame
print("Train Dataset:")
display(train.head(10))
print("Main Dataset Shape:", train.shape[0], "rows and", train.shape[1], "columns\n")

# Display data types & missing values of each column in df_train
print("Data Types & Missing Values of Each Column:")
display(train.info())

# Loop through each column and print unique values
for col in train.columns:
    unique_vals = train[col].unique()
    print(f"Column: {col}")
    print(f"Unique values ({len(unique_vals)}): {unique_vals}\n")

print(f"\nData types check:")
for col in train.columns:
    dtype = train[col].dtype
    unique_vals = train[col].nunique()
    print(f"  ‚û°Ô∏è{col:<25} {str(dtype):<10} [{unique_vals}] unique values")
```

```python
# Display df_train DataFrame
print("Test Dataset:")
display(test.head(10))
print("Main Dataset Shape:", test.shape[0], "rows and", test.shape[1], "columns\n")

# Display data types & missing values of each column in df_train
print("Data Types & Missing Values of Each Column:")
display(test.info())
# Loop through each column and print unique values
for col in test.columns:
    unique_vals = test[col].unique()
    print(f"Column: {col}")
    print(f"Unique values ({len(unique_vals)}): {unique_vals}\n")

print(f"\nData types check:")
for col in test.columns:
    dtype = test[col].dtype
    unique_vals = test[col].nunique()
    print(f"  ‚û°Ô∏è{col:<25} {str(dtype):<10} [{unique_vals}] unique values")
```

# Null values

```python
#Printing null values for each column in train dataset
for col in train.columns:
    null_count = train[col].isnull().sum()
    if null_count > 0:
        print(f"{col}: {null_count}")
```

```python
#Printing null values for each column in test dataset
for col in test.columns:
    null_count = test[col].isnull().sum()
    if null_count > 0:
        print(f"{col}: {null_count}")
```
