Data leakage happens when:
- Information from the test set influences training
- Or when a feature contains **future information**
- Or when preprocessing is done **before splitting**

Very important:
- Remove post-event features
- Remove target proxies
- Remove ID-like features
- Check correlation with target

````python
import numpy as np
from sklearn.metrics import mean_squared_error, r2_score

def evaluate_model(model, X_train, y_train, X_valid, y_valid, name):

    # Train
    model.fit(X_train, y_train)

    # Predict
    preds = model.predict(X_valid)

    # Metrics
    mse = mean_squared_error(y_valid, preds)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_valid, preds)

    print(f"{name}")
    print(f"RMSE: {rmse:.4f}")
    print(f"R2:   {r2:.4f}")
    print("-" * 30)

    return rmse, r2
````

````python
def evaluate_model(model, X_train, y_train, X_valid, y_valid, name):

    model.fit(X_train, y_train)

    preds_log = model.predict(X_valid)

    # Convert back to original scale
    preds = np.expm1(preds_log)
    y_valid_original = np.expm1(y_valid)

    rmse = np.sqrt(mean_squared_error(y_valid_original, preds))
    r2 = r2_score(y_valid_original, preds)

    print(f"{name}")
    print(f"RMSE (original scale): {rmse:.2f}")
    print(f"R2:                    {r2:.4f}")
    print("-" * 30)

    return rmse, r2
````

````python
import os
import numpy as np
import pandas as pd
from scipy import sparse
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# -------------------------------------------------
# 1Ô∏è‚É£ Automatically detect project root
# -------------------------------------------------

CURRENT_DIR = os.getcwd()
PROJECT_ROOT = os.path.abspath(os.path.join(CURRENT_DIR, ".."))

DATA_PATH = os.path.join(PROJECT_ROOT, "data", "processed")
RAW_PATH = os.path.join(PROJECT_ROOT, "data", "raw")

print("Current directory:", CURRENT_DIR)
print("Loading data from:", DATA_PATH)

# -------------------------------------------------
# 2Ô∏è‚É£ Load sparse feature matrices
# -------------------------------------------------

X_train = sparse.load_npz(os.path.join(DATA_PATH, "X_train.npz"))
X_valid = sparse.load_npz(os.path.join(DATA_PATH, "X_valid.npz"))
X_test  = sparse.load_npz(os.path.join(DATA_PATH, "X_test.npz"))

# -------------------------------------------------
# 3Ô∏è‚É£ Load target variables (CSV ‚Äî matches your setup)
# -------------------------------------------------

y_train = pd.read_csv(os.path.join(DATA_PATH, "y_train.csv")).values.ravel()
y_valid = pd.read_csv(os.path.join(DATA_PATH, "y_valid.csv")).values.ravel()

# -------------------------------------------------
# 4Ô∏è‚É£ Log transform (Kaggle scoring metric)
# -------------------------------------------------

y_train_log = np.log1p(y_train)
y_valid_log = np.log1p(y_valid)

# -------------------------------------------------
# 5Ô∏è‚É£ Train Linear Regression
# -------------------------------------------------

model = LinearRegression()
model.fit(X_train, y_train_log)

print("Model trained successfully!")

# -------------------------------------------------
# 6Ô∏è‚É£ Validate model
# -------------------------------------------------

pred_valid_log = model.predict(X_valid)

rmse_log = np.sqrt(mean_squared_error(y_valid_log, pred_valid_log))

pred_valid = np.expm1(pred_valid_log)

rmse_raw = np.sqrt(mean_squared_error(y_valid, pred_valid))
mae = mean_absolute_error(y_valid, pred_valid)
r2 = r2_score(y_valid, pred_valid)

print("\n===== Linear Regression Baseline =====")
print(f"Log RMSE (Kaggle metric): {rmse_log:.5f}")
print(f"Raw RMSE ($): {rmse_raw:,.2f}")
print(f"MAE ($): {mae:,.2f}")
print(f"R2: {r2:.4f}")

# -------------------------------------------------
# 7Ô∏è‚É£ Create Kaggle submission
# -------------------------------------------------

test_raw = pd.read_csv(os.path.join(RAW_PATH, "test.csv"))
ids = test_raw["Id"]

pred_test_log = model.predict(X_test)
pred_test = np.expm1(pred_test_log)

submission = pd.DataFrame({
    "Id": ids,
    "SalePrice": pred_test
})

submission.to_csv("submission.csv", index=False)

print("\nSubmission file created successfully!")
print(submission.head())
````

````python
import numpy as np
from sklearn.metrics import mean_squared_error, r2_score

def evaluate_model(model, X_train, y_train, X_valid, y_valid, name):

    # Train
    model.fit(X_train, y_train)

    # Predict (stay in log space!)
    preds = model.predict(X_valid)

    # Metrics (log scale)
    rmse = np.sqrt(mean_squared_error(y_valid, preds))
    r2 = r2_score(y_valid, preds)

    print(f"{name}")
    print(f"RMSE (log): {rmse:.4f}")
    print(f"R2:         {r2:.4f}")
    print("-" * 30)

    return rmse, r2
````
